```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def step(x): #주석인가?
    y=x>0
    y=y.astype(np.int)
    return y
```

```python
x=np.arange(-2.0, 2.0, 0.001)
y=step(x)
plt.ylim(-2,2)
plt.plot(x,y)
```

```python
def Sigmoid(x):
    return 1 / (np.exp(-x)+1)
```

```python
def exponential(x):
    return np.exp(x)
```

```python
x=np.arange(-2,2,0.0001)
y=exponential(x)
plt.plot(x,y)
plt.title('exponential')
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5,5,0.0001)
y=relu(x)
plt.plot(x,y)
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.title('Relu')
```

```python
def minus_exponential(x):
    return np.exp(x)*-1
```

```python
x=np.arange(-2,2,0.0001)
y=minus_exponential(x)
plt.plot(x,y)
plt.title('minus exponential')
```

```python
def minus_exponential_recipocal(x):
    return 1/np.exp(x)*-1
                                
```

```python
x=np.arange(-2,2,0.0001)
y=minus_exponential_recipocal(x)
plt.plot(x,y)
plt.title('minus exponential recipocal')
```

```python
x=np.arange(-2,2,0.0001)
y=Sigmoid(x)
plt.plot(x,y)
plt.title('sigmoid')
```
